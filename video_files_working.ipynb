```python
import cv2 
```

```python
##load video

video = cv2.VideoCapture("./mountain.mp4")
```

```python
if video is None:
    print("Empty")
else:
    print("loading success")
```

```python
##diplay video
##ret = captures boolean values true or false 
#frame =  return frames ,as numpy ,ret =true orelse false

while True:
    ret , frame = video.read()
    if not ret:
        break

    cv2.imshow("videoFrame" , frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# T0 --> f1
##wait 1ms 
##T1 --> f2
##cv2.waitkey return -1 if not pressed


video.release()
cv2.destroyAllWindows()
```

```python
##write  a video in local storage
##load video

video = cv2.VideoCapture("./mountain.mp4")
##video capture object

fourcc = cv2.VideoWriter_fourcc(*"mp4v")

width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))

##video writer object
out = cv2.VideoWriter("./output.mp4",fourcc, 30, (width,height))

while True:
    ret,frame = video.read()
    if not ret:
        break
    out.write(frame)

##release object at end

out.release()
video.release()
cv2.destroyAllWindows()
```
