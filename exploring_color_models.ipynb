```python
import cv2
```

```python
bgr_image = cv2.imread("./mountain.jpg")
```

```python
if bgr_image is None:
    print("Failed")
else:
    print("Success reading")
```

```python
##convert to grey scale

gray_image = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("gray",gray_image)
cv2.waitKey(4000)
cv2.destroyAllWindows()
```

```python
cv2.imwrite("./gray_image.jpg",gray_image)
```

```python
##convert to hSV
##calcualation and thresholding
gray_image = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2HSV)
cv2.imshow("gray",gray_image)
cv2.waitKey(4000)
cv2.destroyAllWindows()
```

```python
##convert to lab
##calcualation and thresholding
gray_image = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2Lab)
cv2.imshow("gray",gray_image)
cv2.waitKey(4000)
cv2.destroyAllWindows()
```

```python
##convert to RGB

gray_image = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2RGB)
cv2.imshow("gray",gray_image)
cv2.waitKey(4000)
cv2.destroyAllWindows()
```
