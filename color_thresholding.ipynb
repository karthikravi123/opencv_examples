```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./mountain.jpg")
```

```python
plt.imshow(image)
```

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
image_hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(image_hsv)
```

```python
plt.imshow(image_rgb)
```

```python
lower_white = np.array([0,0,190])
upper_white = np.array([30,80,255])
```

```python
##inrange provide with a mask output
mask = cv2.inRange(image_hsv,lower_white,upper_white)
```

```python
plt.imshow(mask)
```

unsigned integer

TYpe            bit width       value Range             Use ccase
np.uint8        8bit            0 to 255                Image Processing (pixel values)
np.uint16       16bit           0 to 65535              High-res image data, sensors
np.uint32       32 bit          0 to ~4.3 billion       Large indices ,memory-efficent
np.uint64       64bit           0 to ~18 quintillion    Big data ,scientific computing

If working with image data,
np.uint8 is your go-too -it's mempry -efficient and matches the 0-255 pixel range perfectly


```python
##huw -180,s-255 ,b -0
black_hsv = np.uint8([[[120,255,0]]])

##to get 120 used indexed
black_bgr = cv2.cvtColor(black_hsv,cv2.COLOR_HSV2BGR)[0][0]
```

```python
##copy original and so here we make computation

image_result= image.copy()
```

```python
image_result[mask > 0]  = black_bgr
```

```python
image_result_rgb = cv2.cvtColor(image_result,cv2.COLOR_BGR2RGB)
```

```python
##plot 

fig, axs = plt.subplots(1,3, figsize = (12,6))
axs[0].imshow(image_rgb)
axs[0].set_title('original Image')
axs[0].axis('off')

axs[1].imshow(mask)
axs[1].set_title('Binary Mask')
axs[1].axis('off')

axs[2].imshow(image_result_rgb)
axs[2].set_title('output Image')
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
