```python
import cv2
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create empty canvas

height ,width = 500,700
##blue color
blue = (255,127,0)

canvas = np.full((height,width,3),blue,dtype=np.uint8)

plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
##draw a line
```

```python
cv2.line(canvas,(50,50),(650,450),(255,255),5)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
#draw a rectangle
#-1 fill whole thing
cv2.rectangle(canvas,(100,100),(300,300),(0,255,255),2)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
#draw a rectangle
#-1 fill whole thing
cv2.rectangle(canvas,(100,100),(300,300),(0,255,255),-1)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
cv2.circle(canvas,(500,150),100,(0,0,255),-1)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
##draw an elipse

cv2.ellipse(canvas,(350,350),(100,50),45,0,360,(0,255,0),3)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
#3shape more than 4 size 

#draw a polygon

#Xn,Yn where n>2

points = np.array([[150,400],[300,400],[225,300]],np.int32)
new_reshaped_points = points.reshape((-1,1,2))
```

```python
#
# -1 = auto
```

```python
new_reshaped_points
```

```python
cv2.polylines(canvas,[new_reshaped_points],True,(255,0,0),2)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```
