```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("giraffe.jpg")
```

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
#flip
```

```python
##1 - horizontally
#0 - vertially
# 1 both side
flip_horizontal = cv2.flip(image,1)
flip_vertical = cv2.flip(image,0)
flip_both = cv2.flip(image,-1)


title =  ["original image","horizoontal flip","vertical flip","flip both Area"]
images = [image_rgb,
          cv2.cvtColor(flip_horizontal,cv2.COLOR_BGR2RGB),
          cv2.cvtColor(flip_vertical,cv2.COLOR_BGR2RGB),
          cv2.cvtColor(flip_both,cv2.COLOR_BGR2RGB)
          ]

plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(title[i])
    plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
#Rotate

rotate_90_clock_wise = cv2.rotate(image_rgb,cv2.ROTATE_90_CLOCKWISE)
rotate_90_counter_clock_wise = cv2.rotate(image_rgb,cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180 = cv2.rotate(image_rgb,cv2.ROTATE_180)



title =  ["original image","rotate_90_clock_wise","rotate_90_counter_clock_wise","rotate_180"]
images = [image_rgb,
          rotate_90_clock_wise,
          rotate_90_counter_clock_wise,
          rotate_180
          ]

plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(title[i])
    plt.axis("off")
plt.tight_layout()
plt.show()
```
