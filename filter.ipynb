```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
##0 as gray scale instead of cv2.IMREAD_GRAYSCALE

image = cv2.imread("./filter.jpg",cv2.IMREAD_GRAYSCALE)

plt.imshow(image,cmap='gray')
```

```python
horizontal_kernel = np.array([[-1 , -1 ,-1],
                       [0,0,0],
                       [1,1,1]])

horizontal_filtered_image = cv2.filter2D(image, -1 , horizontal_kernel)
plt.imshow(horizontal_filtered_image,cmap='gray')
```

```python
##extract horizontal lines no diagnoals
horizontal_kernel = np.array([[-1 , -1 ,-1],
                       [10,10,10],
                       [1,1,1]])

horizontal_filtered_image = cv2.filter2D(image, -1 , horizontal_kernel)
plt.imshow(horizontal_filtered_image,cmap='gray')
```

```python
horizontal_kernel = np.array([[-1 , -1 ,-1],
                       [-25,-25,-25],
                       [1,1,1]])

horizontal_filtered_image = cv2.filter2D(image, -1 , horizontal_kernel)
plt.imshow(horizontal_filtered_image,cmap='gray')
```

```python
x=-15

horizontal_kernel = np.array([[-1 , -1 ,-1],
                       [x,x,x],
                       [1,1,1]])


horizontal_filtered_image = cv2.filter2D(image, -1 , horizontal_kernel)
plt.imshow(horizontal_filtered_image,cmap='gray')
```

```python
x=-15

horizontal_kernel = np.array([[-1 , -1 ,-1],
                       [x,x,x],
                       [1,1,1]])


vertical_kernel = np.array([[-1 , x ,1],
                       [-1,x,1],
                       [-1,x,1]])

diagnoal_kernel = np.array([[x , -1 ,-1],
                       [1,x,-1],
                       [1,1,x]])


horizontal_filtered_image = cv2.filter2D(image, -1 , horizontal_kernel)

vertical_kernel_image = cv2.filter2D(image, -1 , vertical_kernel)

diagnoal_kernel_image = cv2.filter2D(image, -1 , diagnoal_kernel)

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title("original image")
plt.imshow(image,cmap='gray')


plt.subplot(2,2,2)
plt.title("originhorizontalal image")
plt.imshow(horizontal_filtered_image,cmap='gray')

plt.subplot(2,2,3)
plt.title("vertical image")
plt.imshow(vertical_kernel_image,cmap='gray')

plt.subplot(2,2,4)
plt.title("diagonal image")
plt.imshow(diagnoal_kernel_image,cmap='gray')
```
